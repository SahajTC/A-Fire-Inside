```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
def prob_a(ra,rb):
   y=(1/(1+10**((rb-ra)/400)))
   return y

def expected_score_a(oldr_a,arr):
  k=1;
  for i in range(0,10):
    k=k+prob_a(arr[i],oldr_a)
  return k

def nr_a(score_a,oldr_a,arr):
  return oldr_a-25*(score_a-(expected_score_a(oldr_a,arr)))
```
